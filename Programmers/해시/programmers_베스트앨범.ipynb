```python
import pandas as pd
def solution(genres, plays):
    df = pd.DataFrame({'genres':genres,
                  'plays':plays})
    
    genres_rank = df.pivot_table(index = 'genres', values = 'plays', aggfunc = 'sum').sort_values(by = 'plays', ascending = False).index
    
    answer = []
    for a in genres_rank:
        try:
            top2 = df[df.genres == a].sort_values(by = 'plays', ascending=False).head(2).index
        except:
            top2 = df[df.genres == a].sort_values(by = 'plays', ascending=False).head(1).index
        answer.extend(top2)

    return answer
```

```python
genres = ["classic", "pop", "classic", "classic", "pop"]
plays = [500, 600, 150, 800, 2500]
```

```python
solution(genres, plays)
```
