```python
from collections import defaultdict
def solution(clothes):
    answer = 1
    d = defaultdict(list)
    for i, j in clothes:
        d[j].append(i)

    for i in d.keys():
        answer = answer * (len(d[i])+1)
    return answer - 1
```

```python
clothes = [["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]]
```

```python
solution(clothes)
```
