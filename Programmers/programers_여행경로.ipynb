```python
def solution(tickets):
    answer = ["ICN"]
    
    def DFS(tickets, start, used_ticket):
        for i in range(len(tickets)):
            if tickets[i][0] == start and i not in used_ticket:
                used_ticket.append(i)
                DFS(tickets, tickets[i][1], used_ticket)
                if len(used_ticket) == len(tickets):
                    return
                else:
                    used_ticket.pop()
                    
    new_tickets = sorted(tickets, key=lambda x: (x[0], x[1]))
    used_ticket = []
    DFS(new_tickets, answer[0], used_ticket)
    
    for ticket_num in used_ticket:
        answer.append(new_tickets[ticket_num][1])
        
    return answer
```

```python
tickets = [["ICN", "JFK"], ["HND", "IAD"], ["JFK", "HND"]]
```

```python
solution(tickets)
```

```python
from collections import defaultdict
def solution(tickets):
    d = defaultdict(list)
    for i,j in tickets:
        d[i].append(j)
    for i in d.keys():
        d[i].sort()
    # print(r)
    s = ["ICN"]
    p = []
    while s:
        print(s)
        q = s[-1]
        if d[q] != []:
            s.append(d[q].pop(0))
        else:
            p.append(s.pop())
    return p[::-1]
```
