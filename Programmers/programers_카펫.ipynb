```python
def solution(brown, yellow):
    answer = []
    total = brown + yellow
    for i in range(3, int(total**(0.5) + 1)):
        if total % i ==0:
            if (i - 2) * ((total // i) -2) == yellow:
                answer.append(total // i)
                answer.append(i)
                break
    return answer
```
