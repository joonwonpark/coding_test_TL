```python
k = 80
dungeons = [[80,20],[50,40],[30,10]]
```

```python
def solution(k, dungeons):
    global visited, answer, n
    answer = -1
    ## dfs
    n = len(dungeons)
    visited = [0] * n
    cnt = 0
    result = dfs(k, cnt, dungeons)
    return result

def dfs(k, cnt, dungeons):
    # print(visited)
    global answer
    answer = max(cnt, answer)
    # print(f'answer : {answer}')
    for i in range(n):
        # print(f'dfs i:{i}')
        if (k >= dungeons[i][0]) and (visited[i] == 0):
            # print(f'k : {k}, dungeons[i][0]:{dungeons[i][0]}')
            visited[i] = 1
            # print(f'cnt:{cnt}')
            dfs(k - dungeons[i][1], cnt + 1, dungeons)
            # print(f'for i:{i}')
            visited[i] = 0
    return answer
```

```python
solution(k, dungeons)
```
