```python
n = int(input())
x0, y0, e0 = map(int, input().split())
wifi_power = []
for i in range(n):
    x, y, e = map(int, input().split())
    d_e = 0
    for j in wifi_power:
        j[3]= max(0, j[3] - max(0, e-(abs(x-j[0])+abs(y-j[1]))))
        d_e += max(0, j[2] - (abs(x-j[0])+abs(y-j[1])))
    c_e = max(0, e0-(abs(x-x0)+abs(y-y0)))
    c_e = max(0, c_e - d_e - e)
    wifi_power.append([x, y, e, c_e])

answer_list = []
for i in wifi_power:
    answer_list.append(i[3])
if max(answer_list) == 0:
    print('IMPOSSIBLE')
else:
    print(max(answer_list))
```
